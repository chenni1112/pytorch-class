```python
# torch.nn.RNNCell( ) 和torch.nn.RNN()
# 区别在于RNNCell() 只能接受序列中单步的输入，且必须传入隐藏状态，
#而RNN 可以接受一个序列的输入，默认会传入全0的隐藏状态，也可以自己申明隐藏状态的传入

#RNN() 里面的参数有
# input_size() 表示输入xt的特征维度
#hidden_size() 表示输出特征的的维度
#num_layers 表示网络的层数
#nonlinearity 表示选用的非线性激活函数，默认tanh
#bias 表示是否使用偏置，默认使用
# batch_first 表示输入数据的形式，默认是False，就是这样的形式，(seq,batch,feature),也就是，(seq,batch,feature) # 也就是将序列长度放在第一位，
#batch 放在第二位
#dropout 表示是否在输出层应用dropout
# bidirectional 表示是否使用双向的rnn，默认是False
# 对于RNNCell ,里面的参数就少很多，input size ，hidden size，bias 以及nonlinearity
```

```python
import torch
from torch.autograd import Variable

from torch import nn
# 定义一个 单位的rnn
rnn_single=nn.RNNCell(input_size=100,hidden_size=200)
# 访问其中的参数
rnn_single.weight_hh
```

```python
# 构造一个序列，长为6，batch 是5，特征是100
x=Variable(torch.randn(6,5,100))  # 这是rnn 的输入格式
x
```

```python
h_t=Variable(torch.zeros(5,200)) # 定义初始的记忆状态
```

```python
# 传入rnn
out=[]
for i in range(6): # 通过循环6 次作用在整个序列上
    h_t=rnn_single(x[i],h_t)
    out.append(h_t)
```

```python
h_t
```

```python
len(out)
```

```python
out[0].shape # 每个输出的维度
```

```python
# 经过了rnn 之后，隐藏状态的值已经被改变了，因为网络记忆了序列中的信息，同时输出6个结果
```

```python
rnn_seq=nn.RNN(100,200)
```

```python
out,h_t =rnn_seq(x) # 使用默认的全 隐藏状态
h_t
```

```python
# 自己定义的初始的隐藏状态
h_0=Variable(torch.randn(1,5,200))# 有三个维度，分别是(num_layer* num_direction,batch,hidden_size)
```

```python
out,h_t=rnn_seq(x,h_0)
```

```python
# 输出的结果也是(seq,batch,feature)
# 一般使用nn.RNN() ，能避免我们手动写循环，非常方便，，同时如果不特别说明，也会使用默认的全0 初始化隐藏状态
```

```python
#（2） LSTM 有nn.LSTMCell() 和nn.LSTM() 两种形式
lstm_seq=nn.LSTM(50,100,num_layers=2) # 输入维度100，输出200，两层
lstm_seq.weight_hh_l0 # 第一层的h_t 权重
```

```python
lstm_input=Variable(torch.randn(10,3,50))# 序列是10，batch 为3，输入维度是50
out,(h,c)=lstm_seq(lstm_input) # 使用默认的全0 状态
# h和c 都是隐藏状态，大小相同，（num_layers*direction,batch,feature）
```

```python
# 可以不使用 默认的隐藏状态，这是需要传入两个张量
h_init=Variable(torch.randn(2,3,100))
c_init=Variable(torch.randn(2,3,100))
out,(h,c)=lstm_seq(lstm_input,(h_init,c_init))
h.shape
```
