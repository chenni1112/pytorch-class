```python
from  PIL import Image
from torchvision import transforms as tfs
```

```python
im=Image.open("./cat.png")
im
```

```python
#随机比例缩放
# 第一个参数是整数的时候，会将更短的变缩放到这个整数的大小，第一个参数是tuple 的时候，
# 图片会直接把宽和高缩放到这个大小，
# 第二个参数表示缩放图片使用的方法，比如最临近方法，或双线差值
```

```python
print(im.size)
new_im=tfs.Resize((100,200))(im)
print(new_im.size)
new_im
```

```python
# 随机位置截取
# 能够截取图片中的局部信息，使得网络接受的输入具有多尺度的特征，能有较好的效果，
#一个是torchvision.transforms.RandomCrop(),传入的参数是截取出来的图片的长和宽，随机截取
#另一个torchvision.transforms.CenterCrop() 会在图片中心进行截取
```

```python
random_im1=tfs.RandomCrop(100)(im) #100*100
random_im1
```

```python
random_im1=tfs.RandomCrop((150,100))(im) #100*100
random_im1
```

```python
random_im1=tfs.CenterCrop((150,100))(im) #100*100
random_im1
```

```python
# 随机的水平和竖直方向翻转
h_filp=tfs.RandomHorizontalFlip()(im)
h_filp
```

```python
v_filp=tfs.RandomVerticalFlip()(im)
v_filp
```

```python
# 随机角度旋转
rot_im=tfs.RandomRotation(45)(im)
rot_im
```

```python
# 亮度 对比度和颜色变哈
bright_im=tfs.ColorJitter(brightness=1)(im) # 第一个参数是亮度的比例，在0-2 之间变化，
#第二个是对比度，第三个是饱和度，第四个是颜色
bright_im
```

```python
# 对比度
contrast_im=tfs.ColorJitter(contrast=1)(im) # 随机从0-2 之间对比度变化，1表示原图
contrast_im
```

```python
# 颜色
color_im=tfs.ColorJitter(hue=0.5)(im) # 随机从-0.5 ~ 0.5 之间的颜色对比
color_im
```

```python
# 联合使用
im_aug=tfs.Compose([
       tfs.Resize(120),
        tfs.RandomHorizontalFlip(),
        tfs.RandomCrop(96),
        tfs.ColorJitter(brightness=0.5,contrast=0.5,hue=0.5)
        
        
        
    ])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nrows=3
ncols=3
figsize=(8,8)
_,figs=plt.subplots(nrows,ncols,figsize=figsize)
for i in range(nrows):
    for j in range(ncols):
        figs[i][j].imshow(im_aug(im))
        figs[i][j].axes.get_xaxis().set_visible(False)
        figs[i][j].axes.get_xaxis().set_visible(False)

plt.show()
 
```
